```python
import gradio as gr
import ollama

def chat_with_ollama(message, history):
    """
    Sends the user message to Ollama and returns the model's response.
    """
    try:
        response = ollama.chat(
            model="gpt-oss:120b-cloud",
            messages=[
                {"role": "user", "content": message}
            ]
        )
        # Append to chat history
        history.append({"role": "user", "content": response['message']['content']})
        return history

    except Exception as e:
        return history + [("Error", f"Unexpected error: {e}")]
```

```python


# Create Gradio Chat Interface
with gr.Blocks() as demo:
    gr.Markdown("## 💬 Chat with Ollama Model")
    chatbot = gr.Chatbot()
    msg = gr.Textbox(placeholder="Type your message here...")
    clear = gr.Button("Clear Chat")

    msg.submit(chat_with_ollama, [msg, chatbot], chatbot)
    clear.click(lambda: None, None, chatbot, queue=False)

# Run the Gradio app
if __name__ == "__main__":
    demo.launch(server_name="0.0.0.0", server_port=7860)
```
