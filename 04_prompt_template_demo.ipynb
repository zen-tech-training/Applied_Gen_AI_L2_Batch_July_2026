```python
from langchain_ollama import ChatOllama
from langchain_core.prompts import PromptTemplate

model = ChatOllama(model="gpt-oss:120b-cloud")
```

```python
# Example 1 - Single input variable
prompt_template = PromptTemplate(
    template="Which is the capital of {country}?",
    input_variables=["country"]
)
prompt = prompt_template.format(country="Sweden")
print(f"Prompt: {prompt}")
response = model.invoke(prompt)
print(response)
```

```python
# Example 2 - Multiple input variable
prompt_template = PromptTemplate(
    template="What is the sum of {x} and {y}?",
    input_variables=["x", "y"]
)
prompt = prompt_template.format(x=5, y=10)
print(f"Prompt: {prompt}")
response = model.invoke(prompt)
print(response)
```

```python
# Example 3 - Mustache input variables

user_input = {"id": 5, "name": 'Tom', "city": 'Pune'}
prompt_template = PromptTemplate(
    template="User details: ID: {id}, Name: {name}, City: {city}",
    input_variables=["id", "name", "city"]
)   
prompt = prompt_template.format(**user_input)
print(f"Prompt: {prompt}")
response = model.invoke(prompt)
print(response)
```
