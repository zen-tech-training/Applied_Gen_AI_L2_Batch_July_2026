```python
import ollama
```

```python
response =ollama.chat(
    model="gpt-oss:120b-cloud",
    messages=[
        {"role": "system", "content": "You are a mathematics teacher"},
        {"role": "user", "content": "Please explain pythagoras theorem"},
        {"role": "assistant", "content": "Please answer quries in 2 lines"}
    ]
)

print(type(response))
print(response['message']['content'])
```
