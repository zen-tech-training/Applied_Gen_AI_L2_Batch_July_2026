```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage
from langchain_ollama import ChatOllama

model = ChatOllama(model="gpt-oss:120b-cloud")
```

```python
chat_prompt_template =ChatPromptTemplate(
    [
        SystemMessage("You are an expert history teacher."),
        HumanMessage("What is the capital of {country}?")
    ]
)
#OR
chat_prompt_template =ChatPromptTemplate(
    [
        ("system", "You are an expert history teacher."),
        ("human", "What is the capital of {country}?")
    ]
)

prompt = chat_prompt_template.invoke({"country": "Sweden"})
print(f"Prompt: {prompt}")
response = model.invoke(prompt)
print(response)
```

```python
chat_history = []
while True:
    user_input = input("Enter the prompt or quit to exit")
    if user_input.lower() == "quit":
        break
    chat_history.append(HumanMessage(user_input))
    chat_prompt_template =ChatPromptTemplate.from_messages(
    [
        ("system", "You are helpful AI assistant"),
        MessagesPlaceholder(variable_name="chat_history"),
        ("human", "{user_input}")
    ]
    )
    messages = chat_prompt_template.format_messages(chat_history=chat_history, user_input=user_input)
    response = model.invoke(messages)
    chat_history.append(AIMessage(response.content))
    print(f"AI: {response}")

print(f"Chat History: {chat_history}")
```
